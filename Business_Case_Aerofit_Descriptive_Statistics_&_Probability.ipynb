```python
#Missing Value Check - 

data.isna().sum(axis = 0)
```

```python
#Fancy function to check NaNs - 

def null_value_check(data):
    column_name = []
    total_na = []
    
    for col in data.columns:
        
        column_name.append(col)
        
        na_count = data[col].isna().sum()
        total_na.append(na_count)

        tempdf = pd.DataFrame({'Column_Name': column_name, 'Total NA':total_na})
    return tempdf  
```

```python
null_value_check(data)
```

### Observations - 

* Null values are not present in this dataset .

```python
#Outlier Detection Using IQR Method - 

''' 
To compute outliers first thing to do is percentiles  P25 , P50, P75 

To get percentile we are going to use percentile funcrion

'''

P_25 = np.percentile(data["Income"], 25 )
print(f" 25th percentile is {P_25} ")


P_50 = np.percentile(data["Income"], 50 )
print(f" 25th percentile is {P_50} ")


P_75 = np.percentile(data["Income"], 75 )
print(f" 75th percentile is {P_75} ")
```

```python
#Lets compute IQR for income variable -

'''
We know that IQR = 75th percentile - 25th percentile

'''

IQR = P_75 - P_25
print(f" IQR is {IQR} ")
```

```python
#Lets figureout what are whiskers for income variable - 

Left_whisker = P_25 - 1.5 * IQR
print(f" Left whisker is {Left_whisker} ")



Right_whisker = P_75 + 1.5 * IQR
print(f" Right whisker is {Right_whisker} ")
```

```python
#Lets visualize outliers using box plot - 

sns.boxplot(data = data["Income"], orient = "h")
```

```python
#Lets find out total number of outliers for income feature - 

len(data["Income"])
```

```python
Total_outliers = len(data[data["Income"] < Left_whisker]) + len(data[data["Income"] > Right_whisker])
print(Total_outliers)
```
